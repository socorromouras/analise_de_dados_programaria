```python
# Fazer uma função q vai receber os dados de "gestor" e de "nivel", ou seja,
# os dados da coluna gestor e da coluna nivel e vai retornar p gente se a
# pessoa é gestora ou se tem um valor de nivel já definido.
# Nomeando a função a ser criada de "preencher_nivel"...

def preencher_nivel(gestor,nivel):
  if gestor == 1:
    return "Pessoa Gestora"
  else:
    return nivel
```

```python
from scipy.stats import chi2_contingency
```

```python
# CORRELAÇÃO VARIÁVEIS DISCRETAS - CATEGÓRICAS
# COEFICIENTE DE CRAMER
# Frequencia que dos dados da tabela se cruzam: CROSS TABLE

### isso aqui foi muito CONFUSO, mas é necessário para que consigamos executar a função abaixo.
### A instrutora sugeriu colocar no chatGPT

def cramer_coeficiente(coluna1, coluna2):
  tabela_cruzada = np.array(pd.crosstab(coluna1, coluna2))
  chi2 = chi2_contingency(tabela_cruzada)[0]
  soma = np.sum(tabela_cruzada)
  mini = min(tabela_cruzada.shape)-1
  cramer = np.sqrt(chi2/(soma*mini))
  return cramer
```

```python
cramer_coeficiente(dados['COR/RACA/ETNIA'], dados['NIVEL DE ENSINO'])
```

```python
# A gente usa GROUP BY para agrupar por Estado e não ficar várias linhas com o mesmo Estado
# O que tinha antes:
#query = '''SELECT Municipios_Brasileiros.Estado, AVG(Municipios_Status.renda) FROM Municipios_Brasileiros
        #INNER JOIN Municipios_Status ON Municipios_Brasileiros.municipio_ID = Municipios_Status.municipio_ID
        #GROUP BY Municipios_Brasileiros.Estado;'''

# Então, voltamos para a nossa query e vamos reescrevê-la:

query = '''SELECT Municipios_Brasileiros.Estado, AVG(Municipios_Status.renda) FROM Municipios_Brasileiros
        INNER JOIN Municipios_Status ON Municipios_Brasileiros.municipio_ID = Municipios_Status.municipio_ID
        WHERE Municipios_Brasileiros. Estado IN ({})
        GROUP BY Municipios_Brasileiros.Estado;'''.format (','.join(['?' for _ in lista_estados]))

```
