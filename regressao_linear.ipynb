```python
# Estou filtrando tudo que seja "Empregado (CLT)""

dados = dados[dados['QUAL SUA SITUAÇÃO ATUAL DE TRABALHO?']== 'Empregado (CLT)']
```

```python
# Após o FILTRO, ao aplicar value_counts tem que aparecer apenas "Empregado(CLT)"

dados['QUAL SUA SITUAÇÃO ATUAL DE TRABALHO?'].value_counts()
```

```python
# Agora vamos olhar a coluna 'COR/RACA/ETNIA':

dados['COR/RACA/ETNIA'].value_counts()
```

```python
# Como os 4 Indígenas listados não são uma amostra representativa, ou seja,
# não podemos dizer que todos os Indígenas ganham como os 4, vamos excluí-lo,
# mesclando com as outras varíaveis de poucos números, com o "Prefiro não informar" e " Outra"

lista_retirar = ['Prefiro não informar','Outra','Indígena' ]
```

```python
# Função IS IN: se está em, ou seja, se está na lista que a gente está querendo retirar

dados = dados[~dados['COR/RACA/ETNIA'].isin(lista_retirar)]
```

```python
# Criar coluna chamada cor "não branca", onde se a coluna de 'COR/RACA/ETNIA' for branca
# essa coluna recebe zero, e caso contrário recebd 1.
# Para isso usaremos APPLY

dados['NAO_BRANCA'] = dados['COR/RACA/ETNIA'].apply(lambda x:1 if x!= 'Branca' else 0)
```

```python
# Para saber todos os tipos da coluna:

dados['QUANTO TEMPO DE EXPERIÊNCIA NA ÁREA DE DADOS VOCÊ TEM?'].value_counts()
```

```python
#Usando EXTRACT, q é um método q usa expressões regulares, que são REGEX, para extrair padrões específicos de strings.
# A gente ta criando uma coluna chamada TEMPO_EXPERIENCIA. O q está indo nessa coluna:
# ta indo que da coluna de quanto tempo de experiencia, quanto tempo de experiência na área de dados tem, a gente tá pegando linha por
# linha por linha com esse extract, pegando a string, o texto, e a gente tá usandoa função extract, que vai extrair o primeiro número(dígito)
# da expressão "de 1 a 2 anos",por exemplo. Faz fazer isso com todos listados acima. A função para na primeira letra após os dígitos.



dados['TEMPO_EXPERIENCIA'] = dados['QUANTO TEMPO DE EXPERIÊNCIA NA ÁREA DE DADOS VOCÊ TEM?'].str.extract(r'(\d+)')
```

```python
# Olhar como ficou:

dados['TEMPO_EXPERIENCIA']
```

```python
# Add o value_counts, entende melhor tudo, ou melhor, a quantidade de linhas pelo tempo de experiência
# e sem as frases da consulta anterior

dados['TEMPO_EXPERIENCIA'].value_counts()
```

```python
# OLHAR/CONFERIR Coluna 'NUMERO DE FUNCIONARIOS'
# Aqui vai me passa a quantidade de registros por cada categoria/divisão encontrados

dados['NUMERO DE FUNCIONARIOS'].value_counts()
```

```python
# Qual a OBSERVAÇÃO: ao aplicar a função REGEX na coluna exibida na consulta anterior, no caso da primeira linha, onde consta 3.000...
# a função iria parar no ponto e retornar algo errado, ou seja, 3 em vez de 3000. O que fazer?
# LIMPAR para que fique "3000" e não "3.000", substituir o ponto, deixando ali vazio e assim...
# não prejudicar a aplicação da função!

# Vamos atrubuir a função construída à própria coluna para não ter erro:
# Lendo a função: substitir o ponto '.' por '' vazio

dados['NUMERO DE FUNCIONARIOS'] = dados['NUMERO DE FUNCIONARIOS'].str.replace('.','')
```

```python
# Conferindo para ver se o ponto '.' sumiu.
# E o RESULTADO é: SIM!

dados['NUMERO DE FUNCIONARIOS'].value_counts()
```

```python
dados['NUMERO DE FUNCIONARIOS'] = dados['NUMERO DE FUNCIONARIOS'].str.extract(r'(\d+)')
```

```python
# Aplicando um value_counts para confeirir se tudo deu certo.
# A tb aplicando um DROPNA para NÃO PULAR OS NULOS
# RESULTADO: não apareceu nenhum

dados['NUMERO DE FUNCIONARIOS'].value_counts(dropna=False)
```

```python
dados['TEMPO_EXPERIENCIA'].value_counts(dropna=False)

# Aparece 127 NUlos, que é fruto do "Não tenho experiência na área de dados"
```

```python
# O que fazer com esses Nan (Nulos)? Colocar que o tempo de experiência é 0 (Zero)
# Para isso, usa-seo FILLNA (preencher os nulos com "0" - zero)

dados['TEMPO_EXPERIENCIA'] = dados['TEMPO_EXPERIENCIA'].fillna(0)
```

```python
# Conferindo:

dados['TEMPO_EXPERIENCIA'].value_counts(dropna=False)
```

# 8.2 - Preparação dos dados (parte 3)

```python
dados.columns
```

```python
# A gente fez as alterações que foram necessárias nas colunas individuais
# agora a gente pode fazer a seleção das colunas q a gente vai utilizar para o modelo mesmo
# as que vão ser os ATRIBUTOS do nosso modelo e em seguida atribuir a tabela "dados".

dados = dados[['IDADE', 'GENERO', 'NAO_BRANCA','TEMPO_EXPERIENCIA','INSATISFACAO', 'SETOR', 'REGIAO ONDE MORA',
       'NIVEL DE ENSINO','NUMERO DE FUNCIONARIOS','SALARIO', 'NOVO_NIVEL' ]]
```

```python
# Ao aplicar a função abaixo, veremos que a tabela dados terá apenas as colunas que selecionamos acima

dados.columns
```

```python
# Agora a gente vai utilizar o get dummies para as colunas de gênero,
# região onde mora, novo nível e setor. Nesse caso, usaremos realmente o get dummies,
# pois essas variáveis não têm uma ordenação. Por exemplo, eu não posso falar que masculino vale mais que feminino.

# E como o modelo vai criar multiplicadores para cada variável,
# não faria sentido ter valores numéricos aqui. A gente poderia substituir por valores,
# mas como estamos usando também uma análise das importâncias, acabaria influenciando na nossa análise final

dados = pd.get_dummies(dados, columns=['GENERO','SETOR','NOVO_NIVEL', 'REGIAO ONDE MORA'], drop_first=True)
```

* Antes de criar o modelo em si, precisamos separar o nosso conjunto de dados em dois.

* Uma parte vai ser o conjunto de treinamento, que é utilizado pelo modelo para aprender mesmo,

* e a outra parte é o conjunto de teste, utilizado para avaliar o desempenho do modelo  para os dados que o modelo ainda não viu.

```python
# Então, primeiro vamos separar os atributos do nosso objetivo. O atributo: setor, gênero e tal.
#O objetivo, que é o salário.

x = dados.drop('SALARIO', axis=1)
y = dados['SALARIO']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
#Essa função tem como objetivo padronizar as características, removendo a
# média e escalando para a variança unitária. Traduzindo, ela vai normalizar os nossos dados.

# É importante fazer isso pois algoritmos de machine learning, como regressão linear e redes neurais
# funcionam melhor quando os dados têm uma escala uniforme

from sklearn.preprocessing import StandardScaler
```

```python
# Então a gente cria um objeto de normalização, scaler, por exemplo,
# a gente coloca standard scaler. A gente criou esse objeto e aí a gente aplica.

scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

# 8.3 - Introdução a regressão linear (parte 1)

```python
from sklearn.linear_model import LinearRegression
```

```python
# Criar OBJETO dessa regressão Linear:

model = LinearRegression()
```

```python
# Treinar:

model.fit(x_train_scaled, y_train)
```

```python
model.predict(x_test_scaled)
```

```python
y_pred = model.predict(x_test_scaled)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
#Erro quadrático médio (MSE):

mse = mean_squared_error(y_test,y_pred)
```

```python
# Valor absurdamente alto, indicando que o modelo não está com predições precisas

mse
```

```python
# Erro médio absoluto (MAE):

mae = mean_absolute_error(y_test,y_pred)
```

```python
mae
```

```python
r2 = r2_score(y_test,y_pred)
```

```python
r2
```
